```r
library(dplyr)
```

```r
# loading the data
acc <- read.csv('./train/accident_train.csv', header = TRUE)
veh <- read.csv('./train/person_train.csv', header = TRUE)
per <- read.csv('./train/vehicle_train.csv', header = TRUE)
```

```r
# loading the testing data
acc_tst <- read.csv('./test/accident_test.csv', header = TRUE)
veh_tst <- read.csv('./test/person_test.csv', header = TRUE)
per_tst <- read.csv('./test/vehicle_test.csv', header = TRUE)
```

```r
# left joining the training dataframes
acc_per <- left_join(acc, per, by = 'ID')
glimpse(acc_per)
acc_per_veh <- left_join(acc_per, veh, by = 'ID')
glimpse(acc_per_veh)
```

```r
# left joining the testing dataframes
acc_per_tst <- left_join(acc_tst, per_tst, by = 'ID')
glimpse(acc_per_tst)
acc_per_veh_tst <- left_join(acc_per_tst, veh_tst, by = 'ID')
glimpse(acc_per_veh)
```

```r
# write csv
write.csv(acc_per_veh, file="all_train_nohash.csv")
write.csv(acc_per_veh_tst, file="all_test_nohash.csv")
```
